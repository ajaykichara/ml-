```python
import  numpy  as  np 
import  pandas  as  pd 

from sklearn.impute import SimpleImputer 

from  sklearn.preprocessing  import OneHotEncoder
from sklearn.preprocessing  import  OrdinalEncoder 
```

```python
df =  pd.read_csv('covid_toy - covid_toy.csv')
```

```python
df.isnull().sum()
```

```python
from  sklearn.model_selection import   train_test_split 
x_train,x_test,y_train,y_test  =  train_test_split(df.drop(columns=['has_covid']),df['has_covid'],test_size=0.2)
x_train 
```

```python
# adding  simple  imputer  to fever column
si  =  SimpleImputer(strategy = "mean") # (strategy = "most_frequent ") use  for  categorical  data  
# simpleimputer  work  missing  values  fill  only  
x_train_fever =  si.fit_transform(x_train[['fever']])
# also  the  test data  
x_test_fever =  si.fit_transform(x_test[['fever']])
x_train_fever.shape 
```

```python
#  ordinalencoding ----->>>cough 
oe =  OrdinalEncoder(categories = [['Mild','Strong']])
x_train_cough =  oe.fit_transform(x_train[['cough']])
# also  the test  data 
x_test_cough =  oe.fit_transform(x_test[['cough']])
x_train.shape
```

```python
#  onehotEncoding ---->>>>> gender ,city  

ohe = OneHotEncoder(drop='first',sparse_output=False)
x_train_gender_city = ohe.fit_transform(x_train[['gender','city']])

# also  the  test  data 
x_test_gender_city =  ohe.fit_transform(x_test[['gender','city']])
x_train_gender_city.shape 
```

```python
# extracting age  
x_train_age =  x_train.drop(columns=['gender','fever','cough','city']).values
# also  the  test  data  
x_test_age =   x_test.drop(columns=['gender','fever','cough','city']).values
```

```python
x_train_age.shape
```

```python
x_train_transformed =  np.concatenate((x_train_age,x_train_fever,x_train_gender_city,x_train_cough),axis=1)
x_train_transformed.shape
```

```python
# columns  transformer  : - 
# formula:- (constant,approach name ,approach  where you apply or columns ka name )
# template bana detay  hay  jis  say  hamra  kam  structure  tare kay  say  ho jaa taa hay(column transformer may )
# set  of  process ek  structured  formate may  pday hay  in column transformer  

from  sklearn.compose import  ColumnTransformer # this  is  how to  import  ColumnTransformer
transformer =  ColumnTransformer(transformers = [
    ('tnf1',SimpleImputer(),['fever'] ),# in a fever column  by the  
# help of si  we v
# fill missing values  by  mean ,median,mode 
    ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']), # by  this  process we encode  our  data 
    ('tnf3',OneHotEncoder(sparse_output=False,drop='first'),['gender','city']),
     ],remainder =  'passthrough')# remainder =  'passthrough'  =>  its  means  rest all  the columns remain same 
```

```python
transformer.fit_transform(x_train).shape   # pass  data frame  in column transformer  
```

```python
import  numpy  as np 
import  pandas as pd  

from  sklearn.preprocessing import OneHotEncoder
from  sklearn.preprocessing  import  OrdinalEncoder 

from sklearn.impute import  SimpleImputer 


```

```python
df  =  pd.read_csv('tips - tips.csv')
df 
```

```python
from sklearn.model_selection import  train_test_split 
x_train,x_test,y_train,y_test = train_test_split(df.drop(columns=['total_bill']),df['total_bill'],test_size=0.2)
x_train
```

```python
# simple imputer no need becase  no missing value 
df.isnull().sum()
```

```python
x_train 
```

```python
# onehotencoder  
ohe =  OneHotEncoder(drop='first',sparse_output  =  False)
x_train_ohe = ohe.fit_transform(x_train[['smoker','sex']])
x_test_ohe  = ohe.fit_transform(x_test[['smoker','sex']])
x_train_ohe.shape
```

```python
# ordinal encoder 
x_train['time'].value_counts()
oe = OrdinalEncoder(categories = [['Sat','Sun','Thur','Fri']
                                  ,['Dinner','Lunch']])
x_train_oe =  oe.fit_transform(x_train[['day','time']])
x_test_oe =  oe.fit_transform(x_test[['day','time']])
x_train_oe.shape
```

```python
# extracting a  
x_train_tip_size =  x_train.drop(columns=['smoker','sex','day','time']).values
x_test_tip_size =  x_test.drop(columns=['smoker','sex','day','time']).values
x_train_tip_size.shape
```

```python
x_train_transformed  = np.concatenate((x_train_ohe,x_train_oe,x_train_tip_size),axis=1)
x_train_transformed.shape 
```

```python
x_test_transformed  = np.concatenate((x_test_ohe,x_test_oe,x_test_tip_size),axis=1)
x_test_transformed.shape 
```

```python
# columns transformer  
from  sklearn.compose import  ColumnTransformer 
transformer_c = ColumnTransformer([
 ('tnf1',OneHotEncoder(drop='first',sparse_output=False),['smoker','sex']),
 ('tnf2',OrdinalEncoder(categories=[['Sat','Sun','Thur','Fri'],['Dinner','Lunch']]),['day','time'])  # subcategories follow by  columns name
],remainder='passthrough')
```

```python
transformer_c.fit_transform(x_train).shape
```
