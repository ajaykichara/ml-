```python
# ordinal encodr 
import numpy as  np  
import  pandas as  pd 
df  =  pd.read_csv('toyota - toyota.csv')
```

```python
cat_col=['model','transmission','fuelType']
numeric_data = df.drop(columns=cat_col)

from  sklearn.preprocessing import OrdinalEncoder   # not  work  with  fully  data fram  in  which  numeric and categorical data present  so we  devided  into 2  parts  categorical +numerical 
                                                   # then  encode  on  categorical columns  and at the end  add  categoriacal_encoded columns with numeric columns and  now we have full  encoded df 
oe =  OrdinalEncoder(categories=[['Verso-S','Urban Cruiser','IQ','Camry','Supra','PROACE VERSO','Land Cruiser','GT86','Hilux','Verso','Avensis','Prius','Corolla','RAV4','C-HR','Auris','Aygo','Yaris']
                                  ,['Other','Manual','Automatic','Semi-Auto'],
                                 ['Other','Diesel','Hybrid','Petrol']]) # we  arrange  as  priority  already  discuses on  uper strting  
s=oe.fit_transform(df[cat_col])
variable13 = pd.DataFrame(s,columns=cat_col)
variable13

full_encoded_df= numeric_data.merge(variable13,left_index=True,right_index=True)
full_encoded_df
```

```python
import  numpy  as  np  
import  pandas  as  pd 

cat=['sex','smoker','day','time']

df  = pd.read_csv('tips - tips.csv')
df.isnull().sum()
e = pd.get_dummies(df,columns=cat)
e=e.astype(int)
e
```

```python
x= e.drop(columns=['total_bill'])
y=e['total_bill']

from  sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test   =  train_test_split(x,y,test_size=0.2,random_state=42)
from  sklearn.preprocessing  import  MinMaxScaler  
ms =  MinMaxScaler()

x_train_ms = ms.fit_transform(x_train)
x_train_new  =  pd.DataFrame(x_train_ms,columns=x_train.columns)
x_train_new

np.round(x_train_new.describe(),2)

```
