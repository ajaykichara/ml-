```python
import numpy  as np 
import  pandas  as  pd 
df  =  pd.read_csv('covid_toy.csv')
df['fever'] =  df['fever'].fillna(df['fever'].mean())
```

```python
df['has_covid'].value_counts()
df
```

```python
cat_col = ['gender','cough','city','has_covid']
df_new = df.drop(columns=cat_col)

from sklearn.preprocessing import OrdinalEncoder 
oe  =  OrdinalEncoder(categories=[['Male','Female'],
                                  ['Mild','Strong'],
                                  ['Kolkata','Bangalore','Delhi','Mumbai'],
                                  ['No','Yes']])

variable1=  oe.fit_transform(df[cat_col])
new_var =  pd.DataFrame(variable1,columns=cat_col)
new_var

full_dataframe = df_new.merge(new_var,left_index = True,right_index=True)
full_dataframe
```

```python
full_dataframe.head()
full_dataframe.info()
```

```python
full_dataframe
```

```python
x=full_dataframe.drop(columns=['has_covid'])
y=full_dataframe['has_covid']

from  sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =  train_test_split(x,y,test_size=0.2,random_state=42)

from sklearn.preprocessing  import StandardScaler
sc  =  StandardScaler()
train_x_sc =  sc.fit_transform(x_train)
new_train_x= pd.DataFrame(train_x_sc,columns=x_train.columns)
np.round(new_train_x.describe(),2)
```
