```python
import sklearn
print(sklearn.__version__)
#  fianlly  ordinal  encoding  how  use  :
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

df  =  pd.read_csv('toyota - toyota.csv')
df.head()
```

```python
df.isnull().sum()
df['fuelType'].value_counts()
```

```python
 # STEP 1️: Select categorical columns
cat_cols = ['model', 'transmission', 'fuelType']

# Remove numeric columns
df_cat = df[cat_cols]
df_num = df.drop(columns=cat_cols)

# STEP 2️: Ordinal Encoding (ONLY categorical data)
oe = OrdinalEncoder(categories=[
    ['Yaris','Aygo','Auris','C-HR','RAV4','Corolla','Prius','Avensis','Verso',
     'Hilux','GT86','Land Cruiser','PROACE VERSO','Supra','Camry','IQ',
     'Urban Cruiser','Verso-S'],
    ['Manual','Automatic','Semi-Auto','Other'],
    ['Petrol','Hybrid','Diesel','Other']
])

encoded_cat = oe.fit_transform(df_cat)

encoded_cat_df = pd.DataFrame(
    encoded_cat,
    columns=cat_cols,
    index=df.index
)

# STEP 3️: Merge encoded categorical + numeric data
final_df = pd.merge(
    encoded_cat_df,
    df_num,
    left_index=True,
    right_index=True
)

print(final_df.head(20))
```
