```python
# normalization  ----->min=0max=1
import  numpy as  np 
import  pandas  as pd  

df =  pd .read_csv("Churn_Modelling - Churn_Modelling.csv")

df.head(2)
```

```python
df.isnull().sum()
df['Gender'].value_counts()
```

```python
df['Gender']=df['Gender'].map({'Male':0,'Female':1})
df['Geography'] = df['Geography'].map({'France':0,'Germany ':1,'Spain':2})

df=  df.drop(columns=['Surname'])

x = df.drop(columns=['Exited'])
y=df['Exited']


from  sklearn.model_selection import  train_test_split
x_train,x_test,y_train,Y_test = train_test_split(x,y,test_size=0.2,random_state=42)

np.round(x_train.describe(),2)

```

```python
# apply normalization
from  sklearn.preprocessing import MinMaxScaler
mc=MinMaxScaler()
x_train_mc=mc.fit_transform(x_train)
x_train_new = pd.DataFrame(x_train_mc,columns=x_train.columns)
np.round(x_train_new.describe(),2)
```

```python
import pandas  as pd  
import numpy as  np 

df  =  pd.read_csv('toyota - toyota.csv')
df.isnull().sum()
df['transmission'].value_counts()
df=df.drop(columns=['model'])
df['fuelType']=df['fuelType'].map({'Petrol':0,'Hybrid':1,'Diesel':2,'Other':3})
df['transmission']=df['transmission'].map({'Manual':0,'Automatic':1,'Semi-Auto':2,'Other':3})


x  =  df.drop(columns=['price'])
y  = df['price']

from sklearn.model_selection  import  train_test_split
x_train,x_test,y_train,y_test =  train_test_split(x,y,test_size=0.2,random_state=42)

from  sklearn.preprocessing import MinMaxScaler
mt=  MinMaxScaler()
x_train_mt = mt.fit_transform(x_train)
x_train_new = pd.DataFrame(x_train_mt,columns=x_train.columns)

np.round(x_train_new.describe(),2)

```
